```python
# =========================================================
# IMPORT LIBRARIES
# =========================================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings("ignore")

print("Libraries Loaded Successfully")
```

```python
df = pd.read_csv(r'C:\Users\Rutik Bhendarkar\Chat Bot 2\Report Analysis\diabetes.csv')
print(df.head())
```

```python
# =========================================================
# DATASET INFORMATION
# =========================================================

print("Dataset Shape:\n")

print(df.shape)

print("\n")

print("Columns:\n")

print(df.columns)

print("\n")

print("Null Values:\n")

print(df.isnull().sum())

print("\n")

print("Statistics:\n")

print(df.describe())
```

```python
# =========================================================
# TARGET DISTRIBUTION
# =========================================================

plt.figure(figsize=(6,4))

sns.countplot(
    x="Outcome",
    data=df
)

plt.title("Diabetes Distribution")

plt.show()
```

```python
# =========================================================
# CORRELATION MATRIX
# =========================================================

plt.figure(figsize=(12,8))

sns.heatmap(

    df.corr(),

    annot=True,

    cmap="coolwarm"

)

plt.title("Correlation Matrix")

plt.show()
```

```python
# =========================================================
# HANDLE INVALID ZEROS
# =========================================================

invalid_columns = [

    "Glucose",

    "BloodPressure",

    "SkinThickness",

    "Insulin",

    "BMI"
]

for col in invalid_columns:

    df[col] = df[col].replace(

        0,

        df[col].median()
    )

print("Invalid Values Handled")
```

```python
# =========================================================
# FEATURES & TARGET
# =========================================================

X = df.drop(
    "Outcome",
    axis=1
)

y = df["Outcome"]

print("Features Prepared")
```

```python
# =========================================================
# REMOVE OUTLIERS USING IQR
# =========================================================

columns = [

    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI",
    "DiabetesPedigreeFunction",
    "Age"
]

for col in columns:

    Q1 = df[col].quantile(0.25)

    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    df = df[
        (df[col] >= lower) &
        (df[col] <= upper)
    ]

print(df.shape)
```

```python
# =========================================================
# FEATURE ENGINEERING
# =========================================================

# BMI AGE RISK

df["BMI_Age_Risk"] = (

    df["BMI"] * df["Age"]
)

# GLUCOSE BMI INTERACTION

df["Glucose_BMI"] = (

    df["Glucose"] * df["BMI"]
)

# INSULIN GLUCOSE RATIO

df["Insulin_Glucose"] = (

    df["Insulin"] /
    (df["Glucose"] + 1)
)

# HYPERTENSION RISK

df["BP_Age"] = (

    df["BloodPressure"] *
    df["Age"]
)

print("Feature Engineering Completed")
```

```python
# =========================================================
# TRAIN TEST SPLIT
# =========================================================

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
print("Train Test Split Completed")
```

```python
# =========================================================
# FEATURE SCALING
# =========================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print("Scaling Completed")
```

```python
rf_model=RandomForestClassifier(n_estimators=300,max_depth=8,random_state=42)

rf_model.fit(X_train,y_train)

rf_pred=rf_model.predict(X_test)

rf_acc = accuracy_score(y_test,rf_pred)

print("Random Forest Accuracy:")

print(round(rf_acc * 100, 2), "%")
```

```python
# =========================================================
# LOGISTIC REGRESSION
# =========================================================

lr_model = LogisticRegression()

lr_model.fit(X_train_scaled,y_train)

lr_pred = lr_model.predict(X_test_scaled)

lr_acc = accuracy_score(y_test,lr_pred)

print("Logistic Regression Accuracy:")

print(round(lr_acc * 100, 2), "%")
```

```python
# =========================================================
# XGBOOST MODEL
# =========================================================

from xgboost import XGBClassifier

xgb_model = XGBClassifier(

    n_estimators=300,

    learning_rate=0.05,

    max_depth=5,

    subsample=0.8,

    colsample_bytree=0.8,

    random_state=42
)

xgb_model.fit(
    X_train,
    y_train
)

xgb_pred = xgb_model.predict(
    X_test
)

xgb_acc = accuracy_score(
    y_test,
    xgb_pred
)

print("XGBoost Accuracy:")

print(round(xgb_acc * 100, 2), "%")
```

```python
# =========================================================
# MODEL COMPARISON
# =========================================================

comparison = pd.DataFrame({

    "Model": ["Random Forest","Logistic Regression"],

    "Accuracy": [rf_acc,lr_acc]
})

print(comparison)
```

### Adding Condition here

```python
# =========================================================
# IMPROVED RULE-BASED MEDICAL ENGINE
# Better conditions + risk score + user-friendly output
# =========================================================

def medical_rule_engine(report):
    findings = []
    risk_score = 0

    def add_finding(parameter, severity, message, points):
        nonlocal risk_score
        findings.append({
            "parameter": parameter,
            "severity": severity,
            "message": message
        })
        risk_score += points

    # -----------------------------------------------------
    # SAFE GET HELPER
    # -----------------------------------------------------
    def get_value(key, default=None):
        return report.get(key, default)

    # -----------------------------------------------------
    # GLUCOSE
    # -----------------------------------------------------
    glucose = get_value("Glucose")

    if glucose is not None:
        if glucose >= 200:
            add_finding(
                "Glucose",
                "CRITICAL",
                "Very high glucose detected. Severe diabetes risk.",
                30
            )
        elif glucose >= 140:
            add_finding(
                "Glucose",
                "HIGH",
                "High glucose level detected. Diabetes risk present.",
                20
            )
        elif glucose >= 100:
            add_finding(
                "Glucose",
                "MODERATE",
                "Prediabetes range glucose detected.",
                10
            )

    # -----------------------------------------------------
    # BMI
    # -----------------------------------------------------
    bmi = get_value("BMI")

    if bmi is not None:
        if bmi >= 35:
            add_finding(
                "BMI",
                "HIGH",
                "Severe obesity risk detected.",
                15
            )
        elif bmi >= 30:
            add_finding(
                "BMI",
                "MODERATE",
                "Obesity detected.",
                10
            )
        elif bmi < 18.5:
            add_finding(
                "BMI",
                "MODERATE",
                "Low BMI detected. Possible undernutrition risk.",
                8
            )

    # -----------------------------------------------------
    # BLOOD PRESSURE
    # -----------------------------------------------------
    bp = get_value("BloodPressure")

    if bp is not None:
        if bp >= 140:
            add_finding(
                "BloodPressure",
                "HIGH",
                "High blood pressure detected.",
                15
            )
        elif bp <= 90:
            add_finding(
                "BloodPressure",
                "MODERATE",
                "Low blood pressure detected.",
                8
            )

    # -----------------------------------------------------
    # AGE
    # -----------------------------------------------------
    age = get_value("Age")

    if age is not None:
        if age >= 60:
            add_finding(
                "Age",
                "MODERATE",
                "Age-related metabolic risk increased.",
                8
            )
        elif age >= 45:
            add_finding(
                "Age",
                "LOW",
                "Middle-age risk factor detected.",
                4
            )

    # -----------------------------------------------------
    # INSULIN
    # -----------------------------------------------------
    insulin = get_value("Insulin")

    if insulin is not None:
        if insulin >= 200:
            add_finding(
                "Insulin",
                "HIGH",
                "Very high insulin level detected.",
                15
            )
        elif insulin >= 100:
            add_finding(
                "Insulin",
                "MODERATE",
                "Elevated insulin level detected.",
                8
            )

    # -----------------------------------------------------
    # SKIN THICKNESS
    # -----------------------------------------------------
    skin_thickness = get_value("SkinThickness")

    if skin_thickness is not None:
        if skin_thickness >= 35:
            add_finding(
                "SkinThickness",
                "MODERATE",
                "High skin thickness may indicate metabolic risk.",
                6
            )

    # -----------------------------------------------------
    # DIABETES PEDIGREE FUNCTION
    # -----------------------------------------------------
    pedigree = get_value("DiabetesPedigreeFunction")

    if pedigree is not None:
        if pedigree >= 1.0:
            add_finding(
                "DiabetesPedigreeFunction",
                "HIGH",
                "Strong family-history related diabetes risk detected.",
                15
            )
        elif pedigree >= 0.5:
            add_finding(
                "DiabetesPedigreeFunction",
                "MODERATE",
                "Family-history related diabetes risk detected.",
                8
            )

    # -----------------------------------------------------
    # PREGNANCIES
    # -----------------------------------------------------
    pregnancies = get_value("Pregnancies")

    if pregnancies is not None:
        if pregnancies >= 8:
            add_finding(
                "Pregnancies",
                "MODERATE",
                "Higher pregnancy count may increase metabolic monitoring needs.",
                6
            )

    # -----------------------------------------------------
    # FINAL RISK LEVEL
    # -----------------------------------------------------
    if risk_score >= 60:
        overall_risk = "CRITICAL"
    elif risk_score >= 40:
        overall_risk = "HIGH"
    elif risk_score >= 20:
        overall_risk = "MODERATE"
    else:
        overall_risk = "LOW"

    # -----------------------------------------------------
    # USER-FRIENDLY SUMMARY
    # -----------------------------------------------------
    if overall_risk == "CRITICAL":
        summary = "Critical health risk detected. Please seek medical attention soon."
    elif overall_risk == "HIGH":
        summary = "High health risk detected. Medical consultation is recommended."
    elif overall_risk == "MODERATE":
        summary = "Moderate health risk detected. Monitoring and lifestyle care may help."
    else:
        summary = "No major risk detected. Your report looks mostly stable."

    # -----------------------------------------------------
    # SIMPLE RECOMMENDATIONS
    # -----------------------------------------------------
    recommendations = []

    if glucose is not None and glucose >= 140:
        recommendations.append("Reduce sugar intake and monitor blood sugar.")
    if bmi is not None and bmi >= 30:
        recommendations.append("Improve diet and add regular exercise.")
    if bp is not None and bp >= 140:
        recommendations.append("Monitor blood pressure and reduce salt intake.")
    if insulin is not None and insulin >= 100:
        recommendations.append("Check insulin and metabolic health with a doctor.")
    if age is not None and age >= 60:
        recommendations.append("Regular health checkups are recommended.")

    if not recommendations:
        recommendations.append("Maintain a healthy lifestyle and continue regular checkups.")

    return {
        "findings": findings,
        "risk_score": risk_score,
        "overall_risk": overall_risk,
        "summary": summary,
        "recommendations": recommendations
    }
```

```python
sample_report = {
    "Glucose": 220,
    "BMI": 36,
    "BloodPressure": 150,
    "Age": 65,
    "Insulin": 180,
    "DiabetesPedigreeFunction": 1.2,
    "SkinThickness": 40
}

result = medical_rule_engine(sample_report)

print("Summary:", result["summary"])
print("Overall Risk:", result["overall_risk"])
print("Risk Score:", result["risk_score"])

print("\nFindings:")
for item in result["findings"]:
    print(item)

print("\nRecommendations:")
for rec in result["recommendations"]:
    print("-", rec)
```

### HUMAN EXPLANATION ENGINE

```python
# =========================================================
# EXPANDED EXPLAINABLE AI ENGINE
# More clarity for the user
# =========================================================

def generate_human_explanation(result):
    lines = []

    # -----------------------------------------------------
    # HEADER
    # -----------------------------------------------------
    lines.append("=== Medical Report Summary ===")
    lines.append(f"Overall Risk Level: {result.get('overall_risk', 'UNKNOWN')}")
    lines.append(result.get("summary", "No summary available."))
    lines.append("")

    # -----------------------------------------------------
    # SIMPLE RISK MEANING
    # -----------------------------------------------------
    risk_meaning = {
        "LOW": "This report does not show major danger signs.",
        "MODERATE": "This report shows some concern and should be monitored.",
        "HIGH": "This report shows important warning signs and needs attention.",
        "CRITICAL": "This report shows serious warning signs and may need urgent care."
    }

    overall_risk = result.get("overall_risk", "UNKNOWN")
    lines.append("What this means:")
    lines.append(
        risk_meaning.get(
            overall_risk,
            "Risk level could not be interpreted clearly."
        )
    )
    lines.append("")

    # -----------------------------------------------------
    # DETAILED FINDINGS
    # -----------------------------------------------------
    lines.append("=== Detailed Medical Findings ===")

    findings = result.get("findings", [])

    if findings:
        for item in findings:
            parameter = item.get("parameter", "Unknown")
            severity = item.get("severity", "Unknown")
            message = item.get("message", "No message available.")

            if severity == "CRITICAL":
                meaning = "This is a serious abnormal value."
            elif severity == "HIGH":
                meaning = "This value is above normal and should be watched."
            elif severity == "MODERATE":
                meaning = "This is mildly abnormal and may need monitoring."
            else:
                meaning = "This value looks less concerning."

            lines.append(f"- {parameter} [{severity}]")
            lines.append(f"  Result: {message}")
            lines.append(f"  Meaning: {meaning}")
            lines.append("")
    else:
        lines.append("No abnormal findings were detected.")
        lines.append("")

    # -----------------------------------------------------
    # NORMAL / LOW-RISK MESSAGE
    # -----------------------------------------------------
    if not findings:
        lines.append("Good news:")
        lines.append("No major abnormal report values were found in the current analysis.")
        lines.append("")

    # -----------------------------------------------------
    # RECOMMENDATIONS
    # -----------------------------------------------------
    lines.append("=== Recommendations ===")

    recommendations = result.get("recommendations", [])

    if recommendations:
        for rec in recommendations:
            lines.append(f"- {rec}")
    else:
        lines.append("- Maintain a healthy lifestyle and continue regular checkups.")

    lines.append("")

    # -----------------------------------------------------
    # RED FLAG / URGENT SECTION
    # -----------------------------------------------------
    urgent_findings = [
        item for item in findings
        if item.get("severity") == "CRITICAL"
    ]

    if urgent_findings:
        lines.append("=== Important Warning ===")
        lines.append("One or more values are in a serious range.")
        lines.append("Please consult a doctor soon or seek urgent medical care if symptoms are severe.")
        lines.append("")

    # -----------------------------------------------------
    # FUTURE RISK FORECAST
    # -----------------------------------------------------
    lines.append("=== Future Health Forecast ===")

    future_risks = []
    findings_text = str(findings).lower()

    if "glucose" in findings_text:
        future_risks.append(
            "High glucose may lead to future diabetes-related complications if not controlled."
        )

    if "bloodpressure" in findings_text or "blood pressure" in findings_text:
        future_risks.append(
            "High blood pressure may increase the future risk of heart disease or stroke."
        )

    if "bmi" in findings_text:
        future_risks.append(
            "High BMI may increase future metabolic and cardiovascular risk."
        )

    if "insulin" in findings_text:
        future_risks.append(
            "Abnormal insulin values may indicate longer-term metabolic imbalance."
        )

    if "creatinine" in findings_text or "kidney" in findings_text:
        future_risks.append(
            "Kidney-related abnormalities may worsen if hydration and treatment are not managed."
        )

    if future_risks:
        for risk in future_risks:
            lines.append(f"- {risk}")
    else:
        lines.append("- No strong future complication pattern detected from the current findings.")

    lines.append("")

    # -----------------------------------------------------
    # QUICK PLAIN LANGUAGE EXPLANATION
    # -----------------------------------------------------
    lines.append("=== Simple Explanation ===")
    lines.append(
        "This report was checked for unusual values. "
        "The system highlights values that may be too high, too low, or risky. "
        "It then converts them into a simple explanation so the user can understand the result easily."
    )
    lines.append("")

    # -----------------------------------------------------
    # FINAL NOTE
    # -----------------------------------------------------
    lines.append("Note:")
    lines.append(
        "This AI analysis is supportive and not a replacement for a professional medical diagnosis."
    )

    return "\n".join(lines)
```

```python
sample_report = {
    "Glucose": 220,
    "BMI": 36,
    "BloodPressure": 150,
    "Age": 65,
    "Insulin": 180,
    "DiabetesPedigreeFunction": 1.2
}

result = medical_rule_engine(sample_report)
explanation = generate_human_explanation(result)

print(explanation)
```

### Summary Engine

```python
# =========================================================
# ENHANCED FINAL MEDICAL SUMMARY ENGINE
# More clarity for the user
# =========================================================

def generate_final_medical_summary(result):
    lines = []

    # -----------------------------------------------------
    # HEADER
    # -----------------------------------------------------
    lines.append("=========== AI MEDICAL REPORT SUMMARY ===========")
    lines.append("")

    # -----------------------------------------------------
    # OVERALL STATUS
    # -----------------------------------------------------
    overall_risk = result.get("overall_risk", "UNKNOWN")
    summary_text = result.get("summary", "No summary available.")

    lines.append(f"Overall Health Risk Level: {overall_risk}")
    lines.append(summary_text)
    lines.append("")

    # -----------------------------------------------------
    # RISK MEANING
    # -----------------------------------------------------
    risk_meaning = {
        "LOW": "This report does not show major danger signs.",
        "MODERATE": "This report shows some concern and should be monitored.",
        "HIGH": "This report shows important warning signs and needs attention.",
        "CRITICAL": "This report shows serious warning signs and may need urgent care."
    }

    lines.append("What this means:")
    lines.append(risk_meaning.get(overall_risk, "Risk level could not be interpreted clearly."))
    lines.append("")

    # -----------------------------------------------------
    # SEVERITY GUIDE
    # -----------------------------------------------------
    lines.append("Severity Guide:")
    lines.append("- LOW: minor or no major concern")
    lines.append("- MODERATE: mild abnormality, monitor closely")
    lines.append("- HIGH: significant abnormality, medical review advised")
    lines.append("- CRITICAL: serious abnormality, urgent care may be needed")
    lines.append("")

    # -----------------------------------------------------
    # ABNORMAL FINDINGS
    # -----------------------------------------------------
    findings = result.get("findings", [])

    lines.append("Abnormal Findings Detected:")

    if findings:
        for item in findings:
            parameter = item.get("parameter", "Unknown")
            severity = item.get("severity", "Unknown")
            message = item.get("message", "No message available.")

            if severity == "CRITICAL":
                meaning = "Serious abnormal value."
            elif severity == "HIGH":
                meaning = "Clearly above normal and should be reviewed."
            elif severity == "MODERATE":
                meaning = "Mildly abnormal and worth monitoring."
            else:
                meaning = "Not highly concerning."

            lines.append(f"- {parameter} [{severity}] → {message}")
            lines.append(f"  Meaning: {meaning}")
    else:
        lines.append("- No major abnormal findings detected.")
    lines.append("")

    # -----------------------------------------------------
    # VALUE-BASED EXPLANATION
    # -----------------------------------------------------
    lines.append("Simple Explanation:")
    if findings:
        for item in findings:
            parameter = item.get("parameter", "").lower()
            severity = item.get("severity", "")
            if parameter == "glucose":
                if severity == "CRITICAL":
                    lines.append("- Your blood sugar is very high and may need urgent medical review.")
                elif severity == "HIGH":
                    lines.append("- Your blood sugar is high and should be monitored carefully.")
                elif severity == "MODERATE":
                    lines.append("- Your blood sugar is slightly above normal.")
            elif parameter == "bloodpressure":
                if severity == "HIGH":
                    lines.append("- Your blood pressure is high and may increase health risk.")
                elif severity == "MODERATE":
                    lines.append("- Your blood pressure is a little abnormal and should be watched.")
            elif parameter == "bmi":
                if severity == "HIGH":
                    lines.append("- Your BMI suggests high obesity-related risk.")
                elif severity == "MODERATE":
                    lines.append("- Your BMI suggests some weight-related health risk.")
            elif parameter == "insulin":
                if severity in ["HIGH", "CRITICAL"]:
                    lines.append("- Your insulin level is abnormal and may affect metabolism.")
    else:
        lines.append("- No major abnormality was detected in the current report analysis.")
    lines.append("")

    # -----------------------------------------------------
    # FUTURE COMPLICATION ANALYSIS
    # -----------------------------------------------------
    lines.append("Future Health Risk Forecast:")

    findings_text = str(findings).lower()
    future_risks = []

    if "glucose" in findings_text:
        future_risks.append("Future diabetes complications may happen if glucose stays uncontrolled.")
    if "bloodpressure" in findings_text or "blood pressure" in findings_text:
        future_risks.append("High blood pressure may increase future heart disease or stroke risk.")
    if "bmi" in findings_text:
        future_risks.append("High BMI may increase obesity-related metabolic complications.")
    if "insulin" in findings_text:
        future_risks.append("Insulin imbalance may lead to metabolic instability in future.")
    if "creatinine" in findings_text or "kidney" in findings_text:
        future_risks.append("Kidney-related problems may worsen if not managed properly.")

    if future_risks:
        for risk in future_risks:
            lines.append(f"- {risk}")
    else:
        lines.append("- No major future complication pattern detected.")
    lines.append("")

    # -----------------------------------------------------
    # RECOMMENDATIONS
    # -----------------------------------------------------
    lines.append("Medical Recommendations:")
    recommendations = result.get("recommendations", [])
    if recommendations:
        for rec in recommendations:
            lines.append(f"- {rec}")
    else:
        lines.append("- Maintain a healthy lifestyle and continue regular checkups.")
    lines.append("")

    # -----------------------------------------------------
    # NEXT STEPS
    # -----------------------------------------------------
    lines.append("Next Steps:")
    if overall_risk in ["HIGH", "CRITICAL"]:
        lines.append("- Please consult a doctor soon.")
        lines.append("- Follow the recommended diet and lifestyle care.")
        lines.append("- Recheck the report values after treatment or monitoring.")
    elif overall_risk == "MODERATE":
        lines.append("- Monitor your health closely.")
        lines.append("- Improve diet, exercise, and regular checkups.")
    else:
        lines.append("- Keep following a healthy routine.")
        lines.append("- Continue regular health monitoring.")
    lines.append("")

    # -----------------------------------------------------
    # EMERGENCY WARNING
    # -----------------------------------------------------
    critical_exists = any(item.get("severity") == "CRITICAL" for item in findings)

    if critical_exists:
        lines.append("⚠ IMPORTANT WARNING:")
        lines.append("One or more report values are critically abnormal.")
        lines.append("Medical consultation is strongly recommended.")
        lines.append("")

    # -----------------------------------------------------
    # FINAL TAKEAWAY
    # -----------------------------------------------------
    lines.append("Final Takeaway:")
    if overall_risk == "CRITICAL":
        lines.append("The report shows serious concern and should not be ignored.")
    elif overall_risk == "HIGH":
        lines.append("The report shows important warning signs that need medical attention.")
    elif overall_risk == "MODERATE":
        lines.append("The report shows mild to moderate concern and should be monitored.")
    else:
        lines.append("The report does not show a major risk pattern right now.")

    lines.append("")
    lines.append("This AI-generated summary is designed for educational and supportive analysis only.")

    return "\n".join(lines)
```

```python
sample_report = {
    "Glucose": 220,
    "BMI": 36,
    "BloodPressure": 150,
    "Age": 65,
    "Insulin": 180,
    "DiabetesPedigreeFunction": 1.2
}

result = medical_rule_engine(sample_report)
final_summary = generate_final_medical_summary(result)
print(final_summary)
```
