# Healthcare AI Chatbot — Improved Notebook

This notebook keeps the useful features and adds the full improvement path: EDA, preprocessing, feature engineering, model training, top-3 prediction, confidence gap, follow-up questions, emergency detection, explainable AI, and saving artifacts.

```python
# IMPORTS
import os, re, json, warnings
from pathlib import Path
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
from scipy.sparse import hstack

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
```

```python
# LOAD DATASET
CSV_PATH = 'healthcare_dataset.csv'  # change if needed
df = pd.read_csv(CSV_PATH)
print('Shape:', df.shape)
df.head()
```

```python
# AUTO-DETECT TEXT / LABEL COLUMNS
possible_text_cols = ['text', 'symptoms', 'input', 'sentence', 'description', 'query']
possible_label_cols = ['label', 'disease', 'condition', 'class', 'target']

TEXT_COLUMN = next((c for c in possible_text_cols if c in df.columns), None)
LABEL_COLUMN = next((c for c in possible_label_cols if c in df.columns), None)

print('Detected text column:', TEXT_COLUMN)
print('Detected label column:', LABEL_COLUMN)

if TEXT_COLUMN is None or LABEL_COLUMN is None:
    raise ValueError('Could not detect text/label columns. Please set them manually.')
```

## EDA

```python
print(df.info())
print('\nMissing values:\n', df[[TEXT_COLUMN, LABEL_COLUMN]].isnull().sum())
print('\nDuplicate rows:', df.duplicated().sum())
print('\nClass distribution:\n', df[LABEL_COLUMN].value_counts())
```

## Cleaning and preprocessing

```python
df = df.dropna(subset=[TEXT_COLUMN, LABEL_COLUMN]).copy()
df[TEXT_COLUMN] = df[TEXT_COLUMN].astype(str)
df[LABEL_COLUMN] = df[LABEL_COLUMN].astype(str)
df = df.drop_duplicates(subset=[TEXT_COLUMN, LABEL_COLUMN]).copy()
df = df[df[TEXT_COLUMN].str.strip().ne('')]
df = df[df[LABEL_COLUMN].str.strip().ne('')]
df.reset_index(drop=True, inplace=True)
print('After cleaning:', df.shape)
```

```python
def normalize_label(label):
    label = str(label).strip().lower()
    mapping = {
        'covid': 'COVID-19',
        'covid19': 'COVID-19',
        'flu': 'Flu',
        'common cold': 'Common Cold',
        'uti': 'Urinary Tract Infection',
        'gerd': 'GERD',
        'tb': 'Tuberculosis'
    }
    return mapping.get(label, label.title())

df[LABEL_COLUMN] = df[LABEL_COLUMN].apply(normalize_label)
print(df[LABEL_COLUMN].value_counts().head(20))
```

```python
try:
    from textblob import TextBlob
    HAS_TEXTBLOB = True
except Exception:
    HAS_TEXTBLOB = False

def correct_spelling(text):
    if not HAS_TEXTBLOB:
        return str(text)
    try:
        return str(TextBlob(str(text)).correct())
    except Exception:
        return str(text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = set(stopwords.words('english')) - {'no', 'not', 'with', 'without', 'having', 'against'}
lemmatizer = WordNetLemmatizer()

SYMPTOM_SYNONYMS = {
    'temperature': 'fever',
    'high temperature': 'fever',
    'body ache': 'body pain',
    'body aches': 'body pain',
    'head pain': 'headache',
    'migraine': 'headache',
    'tummy pain': 'stomach pain',
    'stomach ache': 'stomach pain',
    'throwing up': 'vomiting',
    'breathlessness': 'breathing difficulty',
    'shortness of breath': 'breathing difficulty',
    'runny nose': 'cold',
    'throat pain': 'sore throat',
    'burning urine': 'urinary burning',
    'loose motion': 'diarrhea',
}

def normalize_medical_terms(text):
    text = str(text).lower()
    for key, value in SYMPTOM_SYNONYMS.items():
        text = re.sub(rf'\b{re.escape(key)}\b', value, text)
    return text

def preprocess_text(text):
    text = str(text)
    text = correct_spelling(text)
    text = text.lower()
    text = normalize_medical_terms(text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    words = [w for w in text.split() if w not in stop_words]
    words = [lemmatizer.lemmatize(w) for w in words]
    return ' '.join(words)

df['clean_text'] = df[TEXT_COLUMN].apply(preprocess_text)
df['word_count'] = df['clean_text'].apply(lambda x: len(str(x).split()))
df = df[df['word_count'] >= 2].copy()
df.reset_index(drop=True, inplace=True)
print(df[[TEXT_COLUMN, 'clean_text']].head())
```

```python
EMERGENCY_WORDS = {'chest', 'pain', 'breathing', 'difficulty', 'stroke', 'fainting', 'bleeding', 'unconscious', 'seizure'}
SEVERE_WORDS = {'severe', 'high', 'intense', 'extreme', 'worsening', 'critical'}
DURATION_WORDS = {'day', 'days', 'week', 'weeks', 'month', 'months', 'today', 'yesterday', 'since'}

df['text_length'] = df['clean_text'].apply(len)
df['unique_word_count'] = df['clean_text'].apply(lambda x: len(set(str(x).split())))
df['symptom_density'] = df['unique_word_count'] / (df['word_count'] + 1)
df['emergency_score'] = df['clean_text'].apply(lambda x: sum(1 for w in EMERGENCY_WORDS if w in set(str(x).split())))
df['severity_score'] = df['clean_text'].apply(lambda x: sum(1 for w in SEVERE_WORDS if w in set(str(x).split())))
df['duration_score'] = df['clean_text'].apply(lambda x: sum(1 for w in DURATION_WORDS if w in set(str(x).split())))
print(df[['word_count', 'text_length', 'unique_word_count', 'symptom_density']].describe())
```

## Train / test split

```python
X_text_raw = df['clean_text']
y = df[LABEL_COLUMN]

X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text_raw,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print('Train size:', X_train_text.shape[0])
print('Test size:', X_test_text.shape[0])
print('Overlap count:', len(set(X_train_text).intersection(set(X_test_text))))
```

## Vectorization and models

```python
vectorizer = TfidfVectorizer(
    max_features=15000,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.95,
    sublinear_tf=True
)

X_train_tfidf = vectorizer.fit_transform(X_train_text)
X_test_tfidf = vectorizer.transform(X_test_text)

extra_cols = ['word_count', 'emergency_score', 'severity_score', 'duration_score', 'text_length', 'unique_word_count', 'symptom_density']
X_train_extra = df.loc[X_train_text.index, extra_cols].values
X_test_extra = df.loc[X_test_text.index, extra_cols].values

X_train = hstack([X_train_tfidf, X_train_extra])
X_test = hstack([X_test_tfidf, X_test_extra])

print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
```

```python
lr_model = LogisticRegression(C=2.0, max_iter=5000, solver='lbfgs', class_weight='balanced', random_state=42)
svm_model = LinearSVC(C=1.5, class_weight='balanced', max_iter=5000, random_state=42)
nb_model = MultinomialNB(alpha=0.5)
rf_model = RandomForestClassifier(n_estimators=200, max_depth=30, min_samples_split=5, min_samples_leaf=2, class_weight='balanced', random_state=42, n_jobs=-1)

models = {
    'Logistic Regression': lr_model,
    'Linear SVM': svm_model,
    'Naive Bayes': nb_model,
    'Random Forest': rf_model,
}

results = {}
trained_models = {}

for name, model in models.items():
    print(f'Training {name}...')
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds, average='weighted')
    results[name] = {'accuracy': acc, 'f1_weighted': f1}
    trained_models[name] = model
    print(f'{name} -> Accuracy: {acc:.4f}, F1: {f1:.4f}')

results_df = pd.DataFrame(results).T.sort_values('accuracy', ascending=False)
results_df
```

```python
best_model_name = results_df.index[0]
best_model = trained_models[best_model_name]
print('Best model:', best_model_name)

best_preds = best_model.predict(X_test)
print(classification_report(y_test, best_preds))

cm = confusion_matrix(y_test, best_preds)
plt.figure(figsize=(14, 10))
sns.heatmap(cm, cmap='Blues', cbar=True)
plt.title(f'Confusion Matrix - {best_model_name}')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

## Intelligent helpers

```python
DISEASE_PROFILES = {
    'Flu': {'reason': 'Flu often causes fever, cough, chills, body pain, headache, and weakness.', 'risk': 'MODERATE', 'care': 'Rest well, drink fluids, and monitor fever. See a doctor if symptoms worsen.'},
    'Common Cold': {'reason': 'Common cold usually causes sneezing, runny nose, mild cough, sore throat, and mild fever.', 'risk': 'LOW', 'care': 'Rest, stay hydrated, and use simple symptom relief if needed.'},
    'COVID-19': {'reason': 'COVID-19 may cause fever, cough, breathing issues, fatigue, and loss of smell or taste.', 'risk': 'HIGH', 'care': 'Isolate if needed and seek medical advice if breathing becomes difficult.'},
    'Dengue': {'reason': 'Dengue commonly causes high fever, body pain, headache, weakness, rash, and nausea.', 'risk': 'HIGH', 'care': 'Stay hydrated and seek medical evaluation, especially if fever is high or persistent.'},
    'Malaria': {'reason': 'Malaria often causes fever with chills, sweating, headache, weakness, and body pain.', 'risk': 'HIGH', 'care': 'Seek medical testing and treatment quickly if malaria is suspected.'},
    'Typhoid': {'reason': 'Typhoid usually causes prolonged fever, stomach pain, weakness, nausea, and loss of appetite.', 'risk': 'HIGH', 'care': 'Consult a doctor for proper diagnosis and treatment.'},
    'Pneumonia': {'reason': 'Pneumonia may cause fever, cough, mucus, chest pain, and breathing difficulty.', 'risk': 'CRITICAL', 'care': 'Urgent medical attention is recommended, especially with shortness of breath.'},
}

FOLLOWUP_QUESTIONS = {
    'fever': ['How long have you had the fever?', 'Is the fever high or mild?', 'Do you also have chills or body pain?', 'Did the fever start suddenly or gradually?'],
    'cough': ['Is the cough dry or with mucus?', 'How long have you been coughing?', 'Do you have breathing difficulty?', 'Is the cough worse at night?'],
    'headache': ['Is the headache severe or mild?', 'Do you feel nausea or dizziness?', 'Is the pain on one side or the whole head?', 'Do bright lights make it worse?'],
    'chest pain': ['Is the chest pain sharp or pressure-like?', 'Does the pain spread to your arm or jaw?', 'Do you also have breathing difficulty?', 'Did it start suddenly?'],
    'stomach pain': ['Where exactly is the stomach pain located?', 'Is the pain severe or mild?', 'Do you also have vomiting or diarrhea?', 'Is it worse after eating food?'],
    'breathing difficulty': ['Is breathing difficulty severe or mild?', 'Did it start suddenly?', 'Do you also have chest pain or cough?', 'Do your lips look bluish?'],
}

EMERGENCY_PATTERNS = [
    {'name': 'Cardiac Emergency', 'any_keywords': ['chest pain', 'chest tightness', 'pressure in chest', 'pain in chest'], 'severity': 'CRITICAL', 'message': 'Possible heart-related emergency. Seek immediate medical attention.'},
    {'name': 'Severe Breathing Emergency', 'any_keywords': ['breathing difficulty', 'shortness of breath', 'cannot breathe', 'severe wheezing'], 'severity': 'CRITICAL', 'message': 'Severe breathing difficulty detected. Seek urgent medical help immediately.'},
    {'name': 'Stroke Warning', 'any_keywords': ['slurred speech', 'face drooping', 'one sided weakness', 'sudden weakness', 'stroke'], 'severity': 'CRITICAL', 'message': 'Possible stroke symptoms detected. Call emergency services immediately.'},
]

KNOWN_SYMPTOMS = ['fever', 'high fever', 'low fever', 'cough', 'dry cough', 'productive cough', 'headache', 'migraine', 'body pain', 'muscle pain', 'joint pain', 'stomach pain', 'abdominal pain', 'cramps', 'chest pain', 'chest tightness', 'breathing difficulty', 'shortness of breath', 'wheezing', 'vomiting', 'nausea', 'diarrhea', 'fatigue', 'weakness', 'rash', 'itching', 'sore throat', 'runny nose', 'sneezing', 'dizziness', 'chills', 'sweating', 'dehydration']

def extract_detected_symptoms(user_input):
    text = user_input.lower()
    return list(dict.fromkeys([s for s in KNOWN_SYMPTOMS if s in text]))

def detect_emergency(user_input):
    text = user_input.lower()
    alerts = []
    for pattern in EMERGENCY_PATTERNS:
        if any(k in text for k in pattern.get('any_keywords', [])):
            alerts.append({'emergency': pattern['name'], 'severity': pattern['severity'], 'message': pattern['message']})
    return alerts

def confidence_gap_analysis(top3):
    if not top3 or len(top3) < 2:
        return {'gap': 0, 'confidence_level': 'LOW', 'message': 'Not enough prediction data.'}
    gap = round(top3[0]['confidence'] - top3[1]['confidence'], 2)
    if gap >= 25:
        return {'gap': gap, 'confidence_level': 'HIGH', 'message': 'Prediction confidence is high.'}
    if gap >= 10:
        return {'gap': gap, 'confidence_level': 'MODERATE', 'message': 'Prediction confidence is moderate.'}
    return {'gap': gap, 'confidence_level': 'LOW', 'message': 'Symptoms overlap with multiple conditions. Please provide more details.'}

def explain_prediction(user_input, predicted_disease, confidence_level='MODERATE'):
    symptoms = extract_detected_symptoms(user_input)
    profile = DISEASE_PROFILES.get(predicted_disease, {})
    reason = profile.get('reason', 'The prediction is based on the symptom pattern detected in the input.')
    if confidence_level == 'HIGH':
        confidence_text = 'The model is highly confident because the symptoms closely match the disease pattern.'
    elif confidence_level == 'LOW':
        confidence_text = 'The model confidence is low because the symptoms overlap with multiple conditions.'
    else:
        confidence_text = 'The model confidence is moderate because some symptoms match and some overlap with other diseases.'
    return {'detected_symptoms': symptoms, 'reasoning': reason, 'confidence_explanation': confidence_text}

def calculate_risk_level(confidence_level, emergency_alerts, detected_symptoms, predicted_disease=None):
    if emergency_alerts:
        return 'CRITICAL'
    disease_risk = DISEASE_PROFILES.get(predicted_disease, {}).get('risk', 'LOW') if predicted_disease else 'LOW'
    if disease_risk in ['CRITICAL', 'HIGH']:
        return disease_risk
    if len(detected_symptoms) >= 5:
        return 'HIGH'
    if confidence_level == 'MODERATE':
        return 'MODERATE'
    return 'LOW'

def get_care_recommendation(risk_level, predicted_disease=None):
    if predicted_disease in DISEASE_PROFILES:
        return DISEASE_PROFILES[predicted_disease]['care']
    return {
        'LOW': 'Monitor symptoms, rest properly, and stay hydrated.',
        'MODERATE': 'Monitor symptoms closely and consider consulting a doctor if symptoms worsen.',
        'HIGH': 'Medical consultation is recommended soon. Avoid self-medication.',
        'CRITICAL': 'Seek immediate emergency medical attention.'
    }.get(risk_level, 'Consult a healthcare professional.')

def generate_followup_questions(user_input, predicted_disease=None, max_questions=5):
    questions = []
    text = user_input.lower()
    for symptom, qlist in FOLLOWUP_QUESTIONS.items():
        if symptom in text:
            questions.extend(qlist)
    if predicted_disease in DISEASE_PROFILES:
        questions.append(f'Would you like to share more details so I can refine the prediction for {predicted_disease}?')
    unique_questions = []
    seen = set()
    for q in questions:
        if q not in seen:
            unique_questions.append(q)
            seen.add(q)
    return unique_questions[:max_questions]
```

```python
def get_top3_predictions(user_input):
    cleaned = preprocess_text(user_input)
    vec_text = vectorizer.transform([cleaned])
    extra = np.array([[len(cleaned.split()), 0, 0, 0, len(cleaned), len(set(cleaned.split())), len(set(cleaned.split())) / (len(cleaned.split()) + 1)]])
    feature_vec = hstack([vec_text, extra])
    probs = lr_model.predict_proba(feature_vec)[0]
    classes = lr_model.classes_
    top3_idx = np.argsort(probs)[-3:][::-1]
    top3_probs = probs[top3_idx]
    normalized_probs = (top3_probs / top3_probs.sum()) * 100 if top3_probs.sum() > 0 else top3_probs
    top3 = []
    for i, idx in enumerate(top3_idx):
        top3.append({'disease': classes[idx], 'confidence': round(float(normalized_probs[i]), 2)})
    return top3

def healthcare_ai_engine(user_input):
    top3 = get_top3_predictions(user_input)
    final_prediction = top3[0]['disease'] if top3 else 'Unknown'
    confidence_result = confidence_gap_analysis(top3)
    explanation = explain_prediction(user_input, final_prediction, confidence_result['confidence_level'])
    emergency_alerts = detect_emergency(user_input)
    followup_questions = generate_followup_questions(user_input, final_prediction)
    risk_level = calculate_risk_level(confidence_result['confidence_level'], emergency_alerts, explanation['detected_symptoms'], final_prediction)
    care = get_care_recommendation(risk_level, final_prediction)
    return {
        'final_prediction': final_prediction,
        'top3_predictions': top3,
        'confidence_analysis': confidence_result,
        'detected_symptoms': explanation['detected_symptoms'],
        'reasoning': explanation['reasoning'],
        'confidence_explanation': explanation['confidence_explanation'],
        'emergency_alerts': emergency_alerts,
        'followup_questions': followup_questions,
        'risk_level': risk_level,
        'care_recommendation': care,
    }

healthcare_ai_engine('I have fever cough chest pain and breathing difficulty')
```

```python
os.makedirs('models', exist_ok=True)
joblib.dump(best_model, 'models/best_model.pkl')
joblib.dump(lr_model, 'models/lr_model.pkl')
joblib.dump(svm_model, 'models/svm_model.pkl')
joblib.dump(vectorizer, 'models/vectorizer.pkl')

with open('models/disease_profiles.json', 'w') as f:
    json.dump(DISEASE_PROFILES, f, indent=4)
with open('models/followup_questions.json', 'w') as f:
    json.dump(FOLLOWUP_QUESTIONS, f, indent=4)
with open('models/emergency_patterns.json', 'w') as f:
    json.dump(EMERGENCY_PATTERNS, f, indent=4)

print('Saved models and configuration files successfully.')
```
